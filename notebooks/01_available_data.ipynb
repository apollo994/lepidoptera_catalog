```python
from Bio import Entrez

def search_sra(species_list):
    Entrez.email = "your.email@example.com"
    results = {}

    for species in species_list:
        handle = Entrez.esearch(db="sra", term=f"{species}[ORGN] AND RNA-Seq[STRA]")
        record = Entrez.read(handle)
        ids = record["IdList"]
        results[species] = ids

    return results

species_list = lepidoptera_species
datasets = search_sra(species_list)

for species, ids in datasets.items():
    print(f"Species: {species}")
    for sra_id in ids:
        print(f"  SRA ID: {sra_id}")
```
